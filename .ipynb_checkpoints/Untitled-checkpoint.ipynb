```python
import xgboost as xgb
```

```python
test_df = pd.read_parquet('amex-default-prediction/test.parquet')
```

```python
categorical_columns = ["customer_ID", "S_2", "D_63", "D_64", "D_66", "D_68", "B_30", "B_38", "D_114", "D_116", "D_117", "D_120", "D_126", "target"]
```

```python
train_raw.loc[:, ~train_raw.columns.isin(categorical_columns)]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    train_raw.loc[:, ~train_raw.columns.isin(categorical_columns)],
    train_raw.iloc[:,-1],
    test_size=0.25)
```

```python
amex_metric_mod(y_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
model = xgb.XGBClassifier(n_estimators=10, max_depth=2)
dtest = xgb.DMatrix(X_test[1:100], y_test[1:100])
dtrain = xgb.DMatrix(X_train[1:100], y_train[1:100])
dtest_x = xgb.DMatrix(X_test[1:100])
dtest_y = xgb.DMatrix( y_test[1:100])

# evallist = [(dtest, 'eval')]
evallist=[(dtest_x, dtest_y)]
evallist = [(X_test, y_test)]
model.fit(X_train[1:100], y_train[1:100]) #, eval_set=evallist, eval_metric = mean_absolute_error,  verbose=True) #  
```

```python
pred_y = model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.roc_auc_score(pred_y, y_test)
```

```python
amex_metric_mod(pred_y, y_test)
```
