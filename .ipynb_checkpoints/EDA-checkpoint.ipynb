```python
import numpy as np
import pandas as pd

from collections import defaultdict

import matplotlib.pyplot as plt
from matplotlib import gridspec

from sklearn.manifold import TSNE
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn import metrics
from sklearn.inspection import permutation_importance

from scipy.stats import spearmanr
from scipy.cluster import hierarchy
from scipy.spatial.distance import squareform

import gc
```

```python
test_df = pd.read_parquet('Processed_Data/test.parquet')
train_df = pd.read_parquet('Processed_Data/cleaned_train.parquet')
```

```python
train_df = train_df.sample(n=100000)
test_df = test_df.sample(n=200000)
```

```python
print(f'test df # customers: {test_df.customer_ID.nunique()}, train df # customers: {train_df.customer_ID.nunique()}')
```

```python
df = train_df
def plot_time_series(prefix='D', cols=None, display_ct=32):
    
    if cols is not None and len(cols)==0: cols = None
    if cols is None:
        COLS = train_df.columns[2:-1]
        COLS = np.sort([int(x[2:]) for x in COLS if x[0]==prefix])
        COLS = [f'{prefix}_{x}' for x in COLS]
        print('#'*25)
        print(f'Plotting all {len(COLS)} columns with prefix {prefix}')
        print('#'*25)
    else:
        COLS = [f'{prefix}_{x}' for x in cols]
        print('#'*25)
        print(f'Plotting {len(COLS)} columns with prefix {prefix}')
        print('#'*25)
    
    for c in COLS:
        tmp = df[['customer_ID', 'S_2', c, 'target']].copy()
        tmp2 = tmp.groupby(['customer_ID', 'target'])[['S_2', c]].agg(list).reset_index()
        tmp3 = tmp2.loc[tmp2.target==1]
        tmp4 = tmp2.loc[tmp2.target==0]
        
        # print(tmp3)
        
        spec = gridspec.GridSpec(ncols=2, nrows=1,
                                 width_ratios=[3,1], wspace=0.1,
                                 hspace=0.5, height_ratios=[1])
        fig = plt.figure(figsize=(20,10))
        ax0 = fig.add_subplot(spec[0])
        
        t0 = []; t1 = []
        for k in range(display_ct):
            try:
                row = tmp3.iloc[k]
                ax0.plot(row.S_2, row[c], '-o', color='blue')
                # print(row.S_2row[c])
                t1 += row[c]
                row = tmp4.iloc[k]
                ax0.plot(row.S_2, row[c], '-o', color='orange')
                t0 += row[c]
            except:
                pass
        plt.title(f'Feature {c} (Key: BLUE=DEFAULT, orange=no default)', size=18)
        
        ax1 = fig.add_subplot(spec[1])
        try:
            t = t0+t1; mn = np.nanmin(t); mx = np.nanmax(t)
            if mx==mn:
                mx += 0.01; mn -= 0.01
            bins = np.arange(mn, mx+(mx-mn)/20, (mx-mn)/20)
            
            if np.sum(np.isnan(t1))!=len(t1):
                ax1.hist(t1, bins=bins, orientation='horizontal', alpha=0.8, color='blue')
            if np.sum(np.isnan(t0))!=len(t0):
                ax1.hist(t0, bins=bins, orientation='horizontal', alpha=0.8, color='orange')
        except:
            pass
        plt.show()
```

```python
# Create a combined data set 
# drop customer_ID, S_2 and target
# indicator is 1 if in training set, 0 in test set

train_df_dist = train_df.drop(columns=['customer_ID', 'S_2', 'target'])
train_df_dist['indicator'] = 1
test_df_dist = test_df.drop(columns=['customer_ID', 'S_2'])
test_df_dist['indicator'] = 0
```

```python
combined_df_dist = train_df_dist.append(test_df_dist)
```

```python
combined_df_dist = combined_df_dist.sample(frac=1) # randomize df 
combined_df_cleaned = combined_df_dist.fillna(0) 
combined_df_cleaned = combined_df_cleaned.reset_index()
```

```python
combined_df_cleaned.head()
```

```python
combined_df_cleaned = combined_df_cleaned.drop(columns='index')
# combined_df_cleaned = combined_df_cleaned.drop(columns='R_1')
# combined_df_cleaned = combined_df_cleaned.drop(columns=['S_11', 'D_59', 'D_120'])
```

```python
X = combined_df_cleaned.iloc[:int(0.8*len(combined_df_cleaned)),:-1]
y = combined_df_cleaned.iloc[:int(0.8*len(combined_df_cleaned)),-1]
clf = RandomForestClassifier(max_depth=5, random_state=1)
clf.fit(X, y)
```

```python
X_test = combined_df_cleaned.iloc[int(0.8*len(combined_df_cleaned)):,:-1]
y_test = combined_df_cleaned.iloc[int(0.8*len(combined_df_cleaned)):,-1]
y_pred = clf.predict(combined_df_cleaned.iloc[int(0.8*len(combined_df_cleaned)):,:-1])
```

```python
roc_auc_score(y_test, y_pred)
```

```python
pd.Series(
    clf.feature_importances_, index=X.columns).sort_values(ascending=True).iloc[-15:].plot.barh()
plt.show()
```

```python
%timeit
perm_array = permutation_importance(clf, X_test, y_test, n_repeats=5, n_jobs=-1)
```

```python
corr = spearmanr(X).correlation
corr = (corr + corr.T)/2
np.fill_diagonal(corr, 1)
distance_matrix = 1 - np.abs(corr)
```

```python
dist_linkage[:,2].max()
```

```python
cluster_ids = hierarchy.fcluster(dist_linkage, 1, criterion='distance')
cluster_id_to_feature_ids = defaultdict(list)

for idx, cluster_id in enumerate(cluster_ids):
    cluster_id_to_feature_ids[cluster_id].append(idx)
selected_features = [v[0] for v in cluster_id_to_feature_ids.values()]

X_sel = X.iloc[:, selected_features]
X_test_sel = X_test.iloc[:, selected_features]

clf_sel = RandomForestClassifier(max_depth=5, random_state=1)
clf_sel.fit(X_test_sel, y_test)

y_pred_sel = clf_sel.predict(X_test_sel)

roc_auc_score(y_test, y_pred_sel)
```

```python
pd.Series(
    clf_sel.feature_importances_, index=X.columns[selected_features]).sort_values(ascending=True).iloc[-15:].plot.barh()
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(30, 30))

dist_linkage = hierarchy.ward(squareform(distance_matrix))
dendro = hierarchy.dendrogram(
    dist_linkage, ax=axes, labels=X.columns.to_list(), leaf_rotation=90)

plt.show()
```

```python
# Summarize data columns
s_list_num = 0
d_list_num = 0
b_list_num = 0
r_list_num = 0
p_list_num = 0

s_list_date = 0

d_list_cat = 0
b_list_cat = 0

date_columns = ['S_2']
categorical_columns = ["D_63", 
                       "D_64", "D_66", 
                       "D_68", "B_30", 
                       "B_38", "D_114", 
                       "D_116", "D_117", 
                       "D_120", "D_126"]  

for i in list(train_df.columns):
    if i in date_columns: s_list_date += 1
    elif i in categorical_columns:
        if i[0] == 'D': d_list_cat+=1
        if i[0] == 'B': b_list_cat+=1
    else:
        if i[0] == 'S': s_list_num+=1
        if i[0] == 'D': d_list_num+=1
        if i[0] == 'B': b_list_num+=1
        if i[0] == 'R': r_list_num+=1
        if i[0] == 'P': p_list_num+=1

assert (len(train_df.columns) - 2) == (s_list_num + d_list_num + b_list_num + r_list_num + p_list_num + s_list_date + d_list_cat + b_list_cat)
print(f'num features: #S : {s_list_num}, #D : {d_list_num}, #B : {b_list_num}, #R : {r_list_num}, #P : {p_list_num}')
print(f'cat features: #D : {d_list_cat}, #B : {b_list_cat}')
print(f'date features: #S : {s_list_date}')
```
