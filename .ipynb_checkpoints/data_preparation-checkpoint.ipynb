```python
import numpy as np
import pandas as pd
```

```python
train_raw = pd.read_parquet('Processed_Data/train.parquet')
labels = pd.read_csv('amex-default-prediction/train_labels.csv')
categorical_features = ['B_30', 'B_38', 
                        'D_63', 'D_64', 
                        'D_66', 'D_68', 
                        'D_114', 'D_116', 
                        'D_117', 'D_120', 'D_126'] # categorical classifications provided by host
```

```python
# Merge labels to train df
train_raw = train_raw.merge(labels, left_on='customer_ID', right_on='customer_ID')


train_raw['S_2'] = pd.to_datetime(train_raw['S_2'])
train_raw[categorical_features] = train_raw[categorical_features].astype("category")
train_raw[categorical_features].dtypes
```

```python
train_raw.describe()
```

```python
train_raw.to_parquet('Processed_Data/cleaned_train.parquet', index=False)
```
